### **STEP 5**

**Handling Missing Values**
> Strategi Mengisi Missing Values:
>
- Tipe Data | Metode | Kapan digunakan
>
- Numerik   | Median | Data memiliki outlier
- Numerik   | Mean   | Data terdistribusi normal
- Kategorik | Modus  | Ambil nilai paling sering
- Kategorik |'Unknown'| Jika tidak ada modus jelas

>

**Isi Missing Values - NUMERIK**

```
# buat copy data
df_clean = df.copy()
```

```
# pilih kolom numerik
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns
```

```
# Isi dengan MEDIAN
for col in numeric_cols:
  if df_clean[col].isnull().sum() > 0:
      median_value = df_clean[col].median()
      df_clean[col].fillna(median_value, inplace=True)
      print(f"{col}: diisi dengan median = {median_value:.2f}")
```

>

**Kenapa MEDIAN ?**
>
- Median tidak terpengaruh outlier
- Lebih aman untuk dengan distribusi skew
- Mean bisa bias jika ada nilai ekstrem

**Isi Missing Values -  KATEGORIK**

```
# Pilih kolom kategorik
categorical_cols = df_clean.select_dtypes(include=['object']).columns
```

```
# Isi dengan MODUS
for col in categorical_cols:
  if  df_clean[col].isnull().sum() > 0:
    mode_value = df_clean[col].mode()[0] if not df_clean[col].mode().empty else 'Unknown'
    df_clean[col].fillna(mode_value, inplace=True)
    print(f"{col}: diisi dengan modus = {mode_value}")
```

>

**Metode Alternatif**

```
# Hapus baris dengan Missing Values
# Hanya jika missing values < 5% dari total data
df_clean.dropna(inplace=True)
```

**Deteksi Outlier**

```
sns.boxplot(data=df_clean, x='Price_in_thousands')
plt.title('Deteksi Outlier dengan Boxplot')
plt.show()
```

```
def deteksi_outlier(df, kolom):
    Q1 = df[kolom].quantile(0.25)
    Q3 = df[kolom].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outlier = df[(df[kolom] < lower) | (df[kolom] > upper)]

    print(f"=== Deteksi Outlier: {kolom} ===")
    print(f"Batas bawah: {lower:.2f}")
    print(f"Batas atas: {upper:.2f}")
    print(f"Jumlah outlier: {len(outlier)} ({len(outlier)/len(df)*100:.2f}%)")
    print(f"\nData outlier:")
    print(outlier[[kolom]])

    return outlier

# Panggil fungsi
outlier_data = deteksi_outlier(df_clean, 'Price_in_thousands')
```

>

**Handling Outlier**

```
# Deteksi outlier dengan IQR method
# Sebelum filter, hitung dulu outlier
Q1 = df_clean['Price_in_thousands'].quantile(0.25)
Q3 = df_clean['Price_in_thousands'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Hitung outlier
outlier_bawah = df_clean[df_clean['Price_in_thousands'] < lower_bound]
outlier_atas = df_clean[df_clean['Price_in_thousands'] > upper_bound]

print(f"=== DETEKSI OUTLIER ===")
print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Batas bawah: {lower_bound}")
print(f"Batas atas: {upper_bound}")
print(f"Jumlah outlier bawah: {len(outlier_bawah)}")
print(f"Jumlah outlier atas: {len(outlier_atas)}")
print(f"Total outlier: {len(outlier_bawah) + len(outlier_atas)}")
```

```
# Tampilkan data outlier
print("\n=== DATA OUTLIER ===")
outliers = df_clean[(df_clean['Price_in_thousands'] < lower_bound) |
                    (df_clean['Price_in_thousands'] > upper_bound)]
print(outliers)
```

```
# Filter outlier
df_clean = df_clean[
    (df_clean['Price_in_thousands'] >= lower_bound) &
    (df_clean['Price_in_thousands'] <= upper_bound)
]

# Verifikasi tidak ada lagi outlier
print(f"\n=== VERIFIKASI ===")
print(f"Min value: {df_clean['Price_in_thousands'].min()}")
print(f"Max value: {df_clean['Price_in_thousands'].max()}")
print(f"Apakah ada outlier? {(df_clean['Price_in_thousands'] < lower_bound).any() or (df_clean['Price_in_thousands'] > upper_bound).any()}")
```

**Encoding Kategorik**

```
# Labeling Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# --- LANGKAH 1: Pastikan 'Vehicle_type' tersedia untuk pengkodean ---
# Jika 'Vehicle_type' TIDAK ada di df_clean DAN pasangannya yang dienkode one-hot juga TIDAK ada di df_clean,
# maka pulihkan 'Vehicle_type' dari df asli untuk memastikan kita memiliki sumber untuk pengkodean.
if 'Vehicle_type' not in df_clean.columns and not any(col in df_clean.columns for col in ['type_Passenger', 'type_SUV']):
    df_clean['Vehicle_type'] = df['Vehicle_type'].str.strip()
    print("Restored 'Vehicle_type' column from original 'df' for encoding purposes.")
elif 'Vehicle_type' not in df_clean.columns and any(col in df_clean.columns for col in ['type_Passenger', 'type_SUV']):
    print("'Vehicle_type' was dropped, but one-hot encoded columns are present. No restoration needed.")
elif 'Vehicle_type' in df_clean.columns:
    print("'Vehicle_type' column is already present for encoding.")


# --- LANGKAH 2: Lakukan Pengkodean Label ---
# Lakukan hanya jika 'Vehicle_type' ada DAN 'Vehicle_type_encoded' TIDAK ada.
if 'Vehicle_type' in df_clean.columns and 'Vehicle_type_encoded' not in df_clean.columns:
    df_clean['Vehicle_type_encoded'] = le.fit_transform(df_clean['Vehicle_type'])
    print("Label encoding applied for 'Vehicle_type'.")
elif 'Vehicle_type_encoded' in df_clean.columns:
    print("Label Encoded column 'Vehicle_type_encoded' already exists. Skipping Label Encoding.")
else:
    print("Skipping Label Encoding: 'Vehicle_type' column not found or already encoded.")


# --- LANGKAH 3: Lakukan One-Hot Encoding ---
# Lakukan hanya jika 'Vehicle_type' ada DAN 'type_Passenger' (atau 'type_SUV') TIDAK ada.
# Kolom Vehicle_type asli akan dihapus oleh pd.get_dummies.
if 'Vehicle_type' in df_clean.columns and not any(col in df_clean.columns for col in ['type_Passenger', 'type_SUV']):
    df_clean = pd.get_dummies(df_clean, columns=['Vehicle_type'], prefix=['type'])
    print("One-Hot Encoding applied for 'Vehicle_type'. Original column dropped.")
elif any(col in df_clean.columns for col in ['type_Passenger', 'type_SUV']):
    print("One-Hot Encoded columns (type_Passenger, type_SUV) already exist. Skipping One-Hot Encoding for 'Vehicle_type'.")
else:
    print("Skipping One-Hot Encoding: 'Vehicle_type' column not found or already one-hot encoded.")
```
